```python
import requests
from textblob import TextBlob


def take_jokes():
    jokes = [] 
    url = "https://v2.jokeapi.dev/joke/Programming?type=single&amount=10"  

    response = requests.get(url)  
    data = response.json()  

    if 'jokes' in data:  
        for joke in data['jokes']:
            jokes.append(joke['joke'])  
    else:  
        jokes.append(data['joke'])  

    return jokes


def calculate_sentiment(text):
    blob = TextBlob(text)  
    return blob.sentiment.polarity  


def main():
    jokes = take_jokes()  

    for a, joke in enumerate(jokes, start=1):  
        print(f"Joke #{a}: {joke}\n")  

        sentiment_value = calculate_sentiment(joke)  
        print(f"Sentiment Value of Joke #{a}: {sentiment_value}\n")  

if __name__ == "__main__":
    main()
```
